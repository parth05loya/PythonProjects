```python
##This function checks whether Player 1 wants to be X or O 
def ask_user_X_or_O():
    print("Welcome to Tic Tac Toe! ")
    user_choice = ''

    #this will ensure that function keeps on asking unless user puts valid response 
    while user_choice not in ['X','O']:
        user_choice = input("What would Player 1 like to be - X or O? : ").upper()
    player_1 = ''
    player_2 = ''
    if user_choice == 'X':
        player_1 = 'X'
        player_2 = 'O'
    else:
        player_1 = 'O'
        player_2 = 'X'
    
    #Function Returns choices of players 
    return(player_1,player_2)

#Ask for position from user where player wants to put the X or O
def ask_board_position(user):
    place = 0 
    while place not in range(1,10):
        place = int(input("Which place do you want to play {} : ".format(user)))
    
    return place 

#This function checks whether game is won or not 
def game_win(board):
     if board[1]=='X' and board[2]=='X' and board[3]=='X':
         print('X won')
         return True
     if board[4]=='X' and board[5]=='X' and board[6]=='X':
         print('X won')
         return True   
     if board[7]=='X' and board[8]=='X' and board[9]=='X':
         print('X won')
         return True
     if board[1]=='X' and board[5]=='X' and board[9]=='X':
         print('X won')
         return True
     if board[7]=='X' and board[5]=='X' and board[3]=='X':
         print('X won')
         return True
     if board[1]=='O' and board[2]=='O' and board[3]=='O':
         print('O won')
         return True
     if board[4]=='O' and board[5]=='O' and board[6]=='O':
         print('O won')
         return True   
     if board[7]=='O' and board[8]=='O' and board[9]=='O':
         print('O won')
         return True
     if board[1]=='O' and board[5]=='O' and board[9]=='O':
         print('O won')
         return True
     if board[7]=='O' and board[5]=='O' and board[3]=='O':
         print('O won')
         return True
         
#This function displays the board
def display_board(board):
    print(board[7] + '|' + board[8] + '|' + board[9])
    print(board[4] + '|' + board[5] + '|' + board[6])
    print(board[1] + '|' + board[2] + '|' + board[3])

#This function switches the players 
def switch_player(current_player, player_1, player_2):
    if current_player == player_1:
        current_player = player_2
    else:
        current_player = player_1
    return current_player

#This function checks whether place is occupied at the board already or not 
def place_occupied(board, place):
    if board[place] != '':
        return True
    else:
        return False 

#This function checks whether board is empty or not 
def board_empty(board):
    for i in range(1,10):
        if board[i] == '':
            return True
    else:
        return False

#This is actual game function 
def game():
    game = True
    # Play untill user wants to keep playing 
    while game:
        player_1,player_2 = ask_user_X_or_O()
        print("Player 1 selected {}".format(player_1))
        print("Player 2 is {}".format(player_2))
        current_player= player_1
        board = [1,'','','','','','','','','']
        
        # Play until board is filled
        while board_empty(board) == True:
            place = 0

            #place at point where position is not already occupied 
            while place_occupied(board,place) == True :
                place = ask_board_position(current_player)
            
            #update the board based on chosen place 
            board[place] = current_player 

            #display the updated board 
            display_board(board)

            #if game is won, we need to break out of it 
            if game_win(board) == True:
                print("Game is Won")
                game = False
                break;
                
            #swtich the current player to next player 
            current_player = switch_player(current_player, player_1, player_2)
        else:

            #Once the board is full If the game is won display else display it tie 
            print("Game is tie")

        #This piece of code ensures that you can ask player to replay 
        response = ''
        while response not in ['Y', 'N']:
            response = input("Do you want to play again? - Y or N").upper()
        if response == 'Y':
            game = True
        else:
            game = False
        
     
```
