```python
import requests

df = pd.DataFrame()

for i in range (1,19):
    url = "https://bhagavad-gita3.p.rapidapi.com/v2/chapters/{}/verses/".format(i)
    
    headers = {
    	"x-rapidapi-key": "c94b65ec46msha848a2c7153ace6p154417jsn1d2b9adc9a21",
    	"x-rapidapi-host": "bhagavad-gita3.p.rapidapi.com"
    }
    
    response = requests.get(url, headers=headers)
    
    temp_df = pd.DataFrame(response.json())
    df = df._append(temp_df, ignore_index = True)
    
```
